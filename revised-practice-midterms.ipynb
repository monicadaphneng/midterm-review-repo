```python
#import pandas and json
import pandas as pd
import json
```

```python
df = pd.read_json("/Users/DELL/Downloads/grocery_product_sales.json")
```

```python
df.columns
```

```python
# load sample data
df.head(10)
```

```python
# check the columns' data type
print(df.dtypes)
```

```python
############################ TASK 1: 1 row = 1 line
```

```python
# convert items column into string for consistent data type and easier manipulation
df['products'] = df['products'].astype(str)
```

```python
# split the items column into individual line items using the ; delimeter
df["products"] = df["products"].str.split(";")
```

```python
df
```

```python
# use the explode method to turn transaction items into individual line items
df = df.explode("products").reset_index(drop=True)
```

```python
# to see if each product is in its own invoice row
print(df[["invoice_id", "products"]])
```

```python
############################ TASK 2: Extract SKU without product quant
```

```python
# extract the SKU by stripping last 5 characters or the quantity, (ex removing x4)
df["SKU"] = df["products"].str[:-5]
```

```python
df
```

```python
############################ TASK 3: Extract Brand
```

```python
# Extract brand (first word before space in SKU column), it is the first word
df["brand"] = df["SKU"].str.split().str[0]
```

```python
df
```

```python
############################ TASK 4: EXTRACT THE QUANTITY
```

```python
# uses regex to find patterns
df["quantity"] = df["products"].str.extract(r"\(x(\d+)\)").astype(int)
```

```python
df
```

```python
############################ TASK 5: EXTRACT PRICE
```

```python
# Step 1: Compute total quantity per invoice, group rows by INVOICE, adds up quantity for each invoice
df["total_qty"] = df.groupby("invoice_id")["quantity"].transform("sum")

# Step 2: Compute price per unit (invoice_total / total_qty)
df["unit_price"] = df["total"] / df["total_qty"]

# Step 3: Compute line subtotal (unit_price * quantity)
df["line_total"] = df["unit_price"] * df["quantity"]

# Clean up if needed (drop helper column)
df = df.drop(columns=["total_qty"])
```
